# Derivatives

- In artificial neural networks (ANNs), derivatives play a crucial role in the training process, specifically in gradient descent optimization.
- The derivative of a function represents the rate of change of that function, and it is used to update the weights and biases of the network during training.

#### The following are the common types of derivatives used in ANNs:

- Weight derivatives: The derivative of the cost function with respect to the weights of the network is used to update the weights during training.

- Bias derivatives: The derivative of the cost function with respect to the biases of the network is used to update the biases during training.

- Activation function derivatives: The derivatives of the activation functions used in the network (such as sigmoid, tanh, or ReLU) are used in the backpropagation algorithm to calculate the gradients of the cost function.

These derivatives are used in the backpropagation algorithm to update the weights and biases of the network in the direction that minimizes the cost function, allowing the network to learn and make more accurate predictions.

# Why bother with Derivatives

- Derivatives point in the direction of increases and decreases
in a mathematical function.
- In DL, the goal (e.g., classification) is represented as an error
function.
- Thus, the best solution is the point with the smallest
error.
- The derivative tells us which way to "move" in that error landscape in order to find the optimal solution.

```python
import numpy as np
import matplotlib.pyplot as plt

# sympy = symbolic math in python
import sympy as sym
from sympy import symbols, Derivative
import sympy.plotting.plot as symplot
```

```python
# create symbolic variables in sympy
x=sym.symbols('x')

# create a function
fx=2*x**2

# compute its derivative
df=sym.diff(fx,x)
# OR
d=Derivative(fx,x)

#print them
print(df)
d.doit() #prints derivative
```

```python
# plot them
symplot(fx,(x,-4,4),title='The function')
plt.show()

symplot(df,(x,-4,4), title='Its derivtive')
plt.show()
```

```python
# repeat with ReLU and sigmoid

# create symbolic variables in sympy
z=sym.symbols('z')

#create symbolic functions
relu=sym.Max(0,z)
sigmoid=1/(1+sym.exp(-z))

# graph the functions
p=symplot(relu,(z,-4,4),label='ReLU',show=False,line_color='blue')
p.extend(symplot(sigmoid,(z,-4,4), label='sigmoid',show=False, line_color="red"))
p.legend=True
p.title='The Functions'
p.show()
```

```python
# graph their derivatives

p=symplot(sym.diff(relu),(z,-4,4),label='df(ReLU)',show=False, line_color='blue')
p.extend(symplot(sym.diff(sigmoid),(z,-4,4),label='df(Sigmoid)',show=False, line_color='red'))
p.legend=True
p.title='The Derivatives'
p.show()
```
