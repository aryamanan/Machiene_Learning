# Standard Error

- The standard error is a measure of the variability or dispersion of the sampling distribution of a statistic.
- It represents the standard deviation of the distribution of the sample means for normally distributed data.
- The standard error provides an estimate of the uncertainty or the variability in a statistic computed from a sample, and it is used to make inferences about the population from the sample data.

- The standard error can be estimated as the standard deviation of the sample divided by the square root of the sample size:

$$\hat{\sigma} = \frac{\sigma}{\sqrt{n}}$$

where:

$\sigma$ is the population standard deviation
$n$ is the sample size
$\hat{\sigma}$ is the estimated standard error of the mean.





# Is Standard Error and Std. Deviation same?

- Standard deviation is a measure of the spread or dispersion of a set of data.
- It is calculated as the square root of the variance, which is the average of the squared deviations from the mean.
- Standard deviation provides information about the amount of variability within a set of data and is a single number that summarizes the spread of the data.

- Standard error, on the other hand, is a measure of the variability of the sampling distribution of a statistic, such as the mean.
- It provides an estimate of the uncertainty or the variability in a statistic computed from a sample, and it is used to make inferences about the population from the sample data.
- Standard error is calculated as the standard deviation of the sample divided by the square root of the sample size.

# T-Test

- A t-test is a statistical hypothesis test in which the test statistic follows a t-distribution under the null hypothesis.
- It is used to determine if there is a significant difference between the means of two groups or sets of data.

#### There are two main types of t-tests: the independent sample t-test and the paired sample t-test.
  - The independent sample t-test is used to compare the means of two independent groups.
  - It tests the null hypothesis that the means of two groups are equal.
  - The paired sample t-test is used to compare the means of two related groups, such as before and after measurements for the same individuals.
  - It tests the null hypothesis that the difference between the means of the two groups is equal to zero.

#### Use-case of T-test
* The t-test is based on the t-statistic, which measures the difference between the sample mean and the population mean in units of the sample standard deviation.
* The t-statistic is then used to calculate the p-value, which represents the probability of observing a t-statistic as extreme or more extreme than the one computed from the sample, if the null hypothesis is true.
* If the p-value is less than a certain significance level, such as 0.05, the null hypothesis can be rejected, and it can be concluded that there is a significant difference between the means of the two groups.

# Formula for T-Test

- The t-test is used to test the null hypothesis that the population mean is equal to a specified value.
The t-statistic for the test is calculated as the difference between the sample mean and the hypothesized population mean, divided by the standard error of the mean:

$$t = \frac{\bar{x} - \mu}{\hat{\sigma}/\sqrt{n}}$$

where:

$\bar{x}$ is the sample mean
$\mu$ is the hypothesized population mean
$\hat{\sigma}$ is the standard error of the mean, estimated as $\hat{\sigma} = \frac{\sigma}{\sqrt{n}}$
$n$ is the sample size.

# T-Test and P-Value relationship

- In a t-test, the p-value is a measure of the strength of the evidence against the null hypothesis.
- It is the probability of observing a t-statistic as extreme as the one computed from the sample, assuming that the null hypothesis is true.
- If the p-value is small (typically below 0.05), it suggests that the results are unlikely to have occurred by chance and provides evidence in favor of rejecting the null hypothesis in favor of an alternative hypothesis.
- If the p-value is large, it means that the results could have easily arisen by chance, and there is not enough evidence to reject the null hypothesis.

- The p-value is used to make a statistical decision about whether the results of a t-test provide sufficient evidence to conclude that there is a significant difference between the means of two groups.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Generate random t-statistics and p-values
np.random.seed(0)
t_statistics = np.random.normal(loc=0, scale=1, size=100)
p_values = stats.t.sf(np.abs(t_statistics), df=100)

# Create a scatter plot
plt.scatter(t_statistics, p_values)
plt.xlabel('t-statistic')
plt.ylabel('p-value')
plt.title('Relationship between t-statistic and p-value')
plt.show()

# P value has to be below 0.05 to statistically significant
```

```python
# Now let's use the t-test to compare models

import scipy.stats as stats
```

```python
# parameters
n1=30
n2=40
mu1=1
mu2=2

# generate the data

data1= mu1+np.random.randn(n1)
data2=mu2+np.random.randn(n2)
```

```python
# plot them

plt.plot(np.zeros(n1),data1,'ro',markerfacecolor='w',markersize=14)
plt.plot(np.ones(n2), data2,'bs', markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],label=['group 1','group 2'])
plt.show()
```

```python
# t-test via stats package

# _ind=indepent samples
t,p=stats.ttest_ind(data1,data2)
print(t)
print(p)
```

```python
# common way to show t-test results in a plot

fig=plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':12}) # change the font size

plt.plot(0+np.random.randn(n1)/15, data1, 'ro', markerfacecolor='w',markersize=14)
plt.plot(1+np.random.randn(n2)/15, data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],label=['Group1','Group2'])
plt.show()
```
